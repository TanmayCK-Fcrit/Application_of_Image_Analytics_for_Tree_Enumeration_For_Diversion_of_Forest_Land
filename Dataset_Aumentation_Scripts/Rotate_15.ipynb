```python
import os
import cv2
import numpy as np

def flip_images_15_degrees(directory):
    for filename in os.listdir(directory):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(directory, filename)
            img = cv2.imread(img_path)
            if img is not None:
                flipped = cv2.flip(img, 1)  # Horizontal flip
                # Add slight rotation
                rows, cols, _ = flipped.shape
                matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 15, 1)
                flipped_rotated = cv2.warpAffine(flipped, matrix, (cols, rows))
                new_filename = f"flipped_15_{filename}"
                cv2.imwrite(os.path.join(directory, new_filename), flipped_rotated)

input_directory = "C:/ML_Projects/Major Project/DataSet/Species_manual_dataset/Common_Coconut"
flip_images_15_degrees(input_directory)
```
