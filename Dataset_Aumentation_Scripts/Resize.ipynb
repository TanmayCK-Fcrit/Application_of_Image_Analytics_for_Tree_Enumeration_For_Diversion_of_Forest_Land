```python
import os
import cv2

def resize_and_save_images(input_directory, output_directory, width=224
                           , height=224):
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for filename in os.listdir(input_directory):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(input_directory, filename)
            img = cv2.imread(img_path)
            if img is not None:
                resized_img = cv2.resize(img, (width, height))
                new_filename = f"resized_{filename}"
                cv2.imwrite(os.path.join(output_directory, new_filename), resized_img)

input_directory = "C:/ML_Projects/Major Project/DataSet/Species_manual_dataset/Common_Coconut"
output_directory = "C:/ML_Projects/Major Project/DataSet/Species_manual_dataset/Common_Coconut/Resized_640x640"
resize_and_save_images(input_directory, output_directory)
```
