```python
import os
import cv2
import numpy as np
def shear_images(directory, shear_factor=0.2):
    for filename in os.listdir(directory):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(directory, filename)
            img = cv2.imread(img_path)
            if img is not None:
                rows, cols, _ = img.shape
                matrix = np.array([[1, shear_factor, 0], [0, 1, 0]], dtype=float)
                sheared = cv2.warpAffine(img, matrix, (int(cols + shear_factor * rows), rows))
                new_filename = f"sheared_{int(shear_factor*100)}_{filename}"
                cv2.imwrite(os.path.join(directory, new_filename), sheared)

input_directory = "C:/ML_Projects/Major Project/DataSet/Species_manual_dataset/Common_Coconut"
shear_images(input_directory)
```
